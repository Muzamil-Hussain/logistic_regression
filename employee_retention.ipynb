### Employees having 
#### 1. Higher Satisfaction level are less likely to leave the job

#### 3. promoted in last 5 years are less likely to leave the job

```python
pd.crosstab(df.salary,df.left).plot(kind='bar')
```

### Employees with High salary are less likely to leave the job

```python
updated_df = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
updated_df.head()
```

```python
encoded_df = pd.concat([updated_df,pd.get_dummies(updated_df.salary,prefix="salary")],axis='columns')
encoded_df.head(10)
```

```python
encoded_df.drop('salary',axis="columns",inplace=True)
encoded_df.head()
```

```python
X_data = encoded_df
X_data.head(10)
```

```python
y_labels = df.left
y_labels.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data,y_labels,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```
